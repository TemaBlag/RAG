```python
from transformers import pipeline

# Replace this with your own checkpoint
model_checkpoint = "huggingface-course/bert-finetuned-squad"
question_answerer = pipeline("question-answering", model=model_checkpoint)
```

```python
question = "How are you?"
context = "I am fine, thanks"

"""
score: This indicates the confidence level (probability) that the model assigns to the extracted answer.
start: This is the starting character index of the answer in the given context.
end: This is the ending character index of the answer in the given context.
answer: This is the actual answer that the model extracted from the context.
"""

# Use the pipeline to answer the question
result = question_answerer(question=question, context=context)
result['answer']
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
```

```python
input_text = "What is your architechture as NN?"

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt")

# Generate output (text generation)
outputs = model.generate(**inputs)

# Decode the output back into text
output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(output_text)
```

```python
model_checkpoint = "distilgpt2"  # Smaller model
text_generator = pipeline("text-generation", model=model_checkpoint)

# Define your prompt
prompt = "How are you?"

# Generate text
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)
print(generated_text)
```
